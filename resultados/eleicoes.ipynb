```python
import pandas as pd
```

```python
#Ler JSON como DF
df = pd.read_json('election-data-5.json')
#Substituir . por , na coluna perc_elei_2022_opacity
df['perc_elei_2022_opacity'] = df['perc_elei_2022_opacity'].str.replace('.', ',')
#Remover . no meio de números em todo o DataFrame
df = df.applymap(lambda x: x.replace('.', '') if isinstance(x, str) else x)
#Substituir , por . em todo o dataframe
df = df.replace(',', '.', regex=True)
#Simplificar o DF apenas com as colunas que interessam
df = df.reindex(columns=['codigo_uf','nm_uf','perc_elei_2022_opacity'])
#Remover linhas duplicadas, para que cada linha seja um estado
df = df.drop_duplicates(subset=['codigo_uf'], inplace=False)
#Remover a linha em que nm_uf é Brasil
df = df[df.nm_uf != 'Brasil']
#Normalizar a coluna perc_elei_2022_opacity de 0 a 1 em uma nova coluna perc_elei_2022_opacity_norm
df['perc_elei_2022_opacity_norm'] = df['perc_elei_2022_opacity'].apply(lambda x: float(x) / 100)
#Arredondar essa coluna para 2 casas decimais
df['perc_elei_2022_opacity_norm'] = df['perc_elei_2022_opacity_norm'].round(2)
#DF para JSON sem caracteres ASCII
df.to_json('estados.json', orient='records', force_ascii=False)
#Mostrar DF para conferência
df
```
